## Task I: Clustering

(1) Use the kmeans algorithm to cluster the timeseries data. The
implementation of kmeans for sklearn is described here:
http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

You need to make some choices: 
  - Preprocessing: should you scale and center data before clustering?
  - Initialization: How will you choose starting centroids
  - Evaluation: How will you determine the number of clusters to use for next steps of analysis. Here are options implemented in sklearn: http://scikit-learn.org/stable/modules/clustering.html#clustering-performance-evaluation. Here is an implementation of the gap statistic method shown in lecture (use at your discretion): https://github.com/Zelazny7/gap-statistic/blob/master/gap.py
  
Make sure to include your choices for the above in your writeup.

```python
# clustering code goes here
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score

# grab useful data
X = afford_df.drop(columns=['RegionName', 'SizeRank', 'RegionID']).values

# find clusters
nclusters = range(2,51)
dbscore = []
sscore = []

for ncluster in nclusters:
    km = KMeans(n_clusters=ncluster,init='k-means++',n_jobs=2)
    km.fit(X)
    dbscore.append(davies_bouldin_score(X, km.labels_))
    sscore.append(silhouette_score(X, km.labels_,metric='euclidean'))
    
plt.plot(nclusters,dbscore,marker='.',linestyle='-', label=('Davies-Bouldin Score'))
plt.plot(nclusters,sscore,marker='.',linestyle='-', label=('Silhouette Score'))
plt.title('Scoring method for choose # of clusters')
plt.xlabel('Clusters')
plt.ylabel('Score')
plt.legend();
```

_Clustering discussion goes here_

The affordability index, which is our feature here, seems to already be scaled/normalized so there is no need to rescale or center data here. We also don't need to center the data for clustering. However, when we do PCA analysis we will have to center the data there.

There are very few scoring methods that do not really on truth labels for the datasets. Two of these are the Silhouette coefficient and Davies-Bouldin index which give measures of cluster densities and separations. For the Silhouette coefficient a value closer to 1 is better and for Davies-Bouldin index a value closer to 0 is better. For both metrics a cluster size of 3 gave the best results. This includes using a k-mean++ initial cluster positions as well as random starting positions.

Three clusters was the ideal choose based on these two metrics.

(2) Further analysis: Plot the resulting cluster centroids and provide a qualitative description of each cluster. Look at the region names and size ranks for the time series you obtain, do resulting clusters reflect geographic or population similarities across regions?

```python
# clustering analysis goes here
# use best choose for clusters
ncluster=3
km = KMeans(n_clusters=ncluster,init='k-means++',n_jobs=2)
km.fit(X)

# plot results
plt.figure(figsize=(15,ncluster))
colors = ['C0','C1','C2'] # 3 colors for each cluster
for Xdata,label in zip(X,km.labels_):
    ax=plt.subplot(1,3,label+1,label=label)
    plt.plot(Xdata,color=colors[label])  
    plt.title('Cluster: '+str(label+1))
    ax.set_xticklabels([])
```

```python
from mpl_toolkits.basemap import Basemap
from geopy.geocoders import Nominatim

# generate US map
fig, ax = plt.subplots(figsize=(10,5))
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
        projection='lcc',lat_1=32,lat_2=45,lon_0=-95,ax=ax)
m.readshapefile('st99_d00', name='states', drawbounds=True)
geolocator = Nominatim()

# define marker colors and sizes
colors = ['C0','C1','C2'] # 3 colors for each cluster
# marker scaling
sizes = afford_df['SizeRank'].values
maxsize = afford_df['SizeRank'].max()
sizes = maxsize+1-sizes # scale so largest pop has biggest number
newsizes = (((sizes - 1) * (15 - 5)) / (maxsize - 1)) + 1 # rescale rank to 5-15 for marker size

# plot points
for i in range(len(afford_df)):
    location = afford_df.iloc[i]['RegionName']
    try:
        loc = geolocator.geocode(location.split('-')[0])
        x, y = m(loc.longitude, loc.latitude)
    except:
        # if city can't be found, use state
        loc = geolocator.geocode(location.split(',')[1][1:])
        x, y = m(loc.longitude, loc.latitude)        
    plt.plot(x,y,marker='o',color=colors[km.labels_[i]],markersize=newsizes[i])
```

_with discussion here_

Results are plotted above on a US map. Each color represents a cluster and marker sizes represent rank size (population). The south and mid west were clumped into a single cluster (blue). A second cluster (green) contains major cities in California (SF, LA, San Diego) + Seattle: these locations have been historically known for having crazy housing affordability problems. The third cluster (orange) seems to have clumped up all remaining location that aren't the mid west/south or California. Population/Rank size seems to have no effect on clustering here.

There was some exploration done with more clusters to see if more defined clusters form on the usa map. This was not the case. Cluster size of 3 gives the most defined clusters on the map.

## Task 2: Dimensionality Reduction via Decomposition
    
(1) Apply one  "global" dimensionality reduction technique, e.g., PCA or Non-negative Matrix Factorization (http://scikit-learn.org/stable/modules/classes.html#module-sklearn.decomposition) to the affordability data. 

```python
# decomposition code here
import numpy as np
from sklearn.decomposition import TruncatedSVD

# remove mean (centering)
X_1 = np.copy(X)
X_1 = X_1 - np.mean(X_1)
    
# fit
svd1 = TruncatedSVD(n_components=2, n_iter=10, random_state=23)
svd1.fit(X_1) # centered
```

(2) Plot the transformed data in two dimensions. Color regions based on clusters found in Task 1. Plot again coloring based on geography or size. Does the data in reduced dimension reflect any of these characteristics?

```python
# plotting code here
plt.figure()
for row,label in zip(X_1,km.labels_):
    plt.scatter(np.dot(row,svd1.components_[0,:]),np.dot(row,svd1.components_[1,:]),color=colors[label])
plt.title('Clustering in SVD space')
plt.ylabel('Component 2')
plt.xlabel('Component 1')

newsizes = (((sizes - 1) * (55 - 5)) / (maxsize - 1)) + 1
plt.figure()
for row,label,size in zip(X_1,km.labels_,newsizes):
    plt.scatter(np.dot(row,svd1.components_[0,:]),np.dot(row,svd1.components_[1,:]),color=colors[label],s=size)
plt.title('Clustering in SVD space (marker size = rank size)')
plt.ylabel('Component 2')
plt.xlabel('Component 1');
```

_decomposition discussion here_

The three clusters are very nicely defined within the component space. Rank size does not seem to make a difference here. The first two components do a good job of linearizing the data, like the theory says.
